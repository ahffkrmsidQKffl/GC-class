- (a) ë‹¤ìŒ í•¨ìˆ˜ì‹ f(n)ì„ êµ¬í˜„í•˜ê³  f(20)ì˜ ê²°ê³¼ ì¶œë ¥ (2ì )


```python
# í•¨ìˆ˜ì‹ f(n) --> recursion(n) ìœ¼ë¡œ ì´ë¦„ ë³€ê²½ë¨
def recursion(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return recursion(n-1) + recursion(n-2)

n=20
print(f'n={n}, recursion(n)={recursion(n)}')  # í•¨ìˆ˜ í˜¸ì¶œ ë° ê²°ê³¼ ì¶œë ¥
```

## Q11. ì •ë ¬ ì•Œê³ ë¦¬ì¦˜ ì‹œê°„ ì¸¡ì •í•˜ê¸° 
ì •ë ¬ ì•Œê³ ë¦¬ì¦˜(Sorting Algorithms)ì„ ì‚¬ìš©í•˜ì—¬ ì•„ë˜ì™€ ê°™ì´ ì •ë ¬ ì‹¤í–‰ì‹œê°„(elapsed time)ì„ ì¸¡ì •í•˜ì—¬ 
ì¶œë ¥í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•˜ì‹œì˜¤. (ë‹¨, ì •ë ¬ì— ì‚¬ìš©ë˜ëŠ” ìˆ˜ëŠ” 0~1000 ì‚¬ì´ì˜ ì„ì˜ì˜ ì •ìˆ˜ 1000ê°œ, ì •ë ¬ì´ ì•ˆëœ ìƒíƒœ)
 
- (a) ë²„ë¸” ì •ë ¬(Bubble Sort) ì•Œê³ ë¦¬ì¦˜ ì‘ì„±í•˜ê³  ì‹¤í–‰ì‹œê°„ ì¶œë ¥ (1ì )
- (b) í€µ ì •ë ¬(Quick Sort) ì•Œê³ ë¦¬ì¦˜ ì‘ì„±í•˜ê³  ì‹¤í–‰ì‹œê°„ ì¶œë ¥ (1ì )
- (c) ê³„ìˆ˜ ì •ë ¬(Counting Sort) ì•Œê³ ë¦¬ì¦˜ ì‘ì„±í•˜ê³  ì‹¤í–‰ì‹œê°„ ì¶œë ¥ (1ì )


```python
import time
import random

# (a).bubble_sort() 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

arr1 = [random.randint(1, 1000) for _ in range(1000)]
print("ì •ë ¬ì „ ë°°ì—´ with Bubble Sort:", arr1)
print("\n")
start1 = time.time()
bubble_sort(arr1)
end1 = time.time()
print("ì •ë ¬ëœ ë°°ì—´ with Bubble Sort:", arr1)
print("\n")
bubble_time = end1-start1
       
# (b).quick_sort() 
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)

arr2 = [random.randint(1, 1000) for _ in range(1000)]
print("ì •ë ¬ì „ ë°°ì—´ with Quick Sort:", arr2)
print("\n")
start2 = time.time()
sorted_arr2 = quick_sort(arr2)
end2 = time.time()
print("ì •ë ¬ëœ ë°°ì—´ with Quick Sort:", sorted_arr2)
print("\n")
quick_time = end2-start2
    

            
# (c).counting_sort
def counting_sort(arr):
    max_val = max(arr)
    count = [0] * (max_val + 1)

    for num in arr:
        count[num] += 1

    sorted_arr = []

    for i in range(len(count)):
        sorted_arr.extend([i] * count[i])

    return sorted_arr

arr3 = [random.randint(1, 1000) for _ in range(1000)]
print("ì •ë ¬ì „ ë°°ì—´ with Counting Sort:", arr3)
print("\n")
start3 = time.time()
sorted_arr3 = counting_sort(arr3)
end3 = time.time()
print("ì •ë ¬ëœ ë°°ì—´ with Counting Sort:", sorted_arr3)
print("\n")
counting_time = end3-start3


# ê²°ê³¼ ì¶œë ¥
print("1.Bubble Sort   elapsed_time :", bubble_time)
print("2.Quick Sort    elapsed_time :", quick_time)
print("3.Counting Sort elapsed_time :", counting_time)
```
