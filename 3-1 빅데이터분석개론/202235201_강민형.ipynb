# 오픈 데이터 설명 : Telco Customer Churn Dataset

## 개요
Telco Customer Churn Dataset은 통신 회사의 고객 이탈(Churn) 여부를 예측하기 위한 데이터셋입니다. 이 데이터셋은 고객의 다양한 특성과 이탈 여부를 포함하고 있으며, 고객 유지 전략을 수립하는 데 유용한 정보를 제공합니다.

## 데이터셋 설명
이 데이터셋은 총 7,043명의 고객 정보를 포함하고 있으며, 다음과 같은 변수를 포함하고 있습니다.

- `customerID`: 고객 고유의 ID
- `gender`: 성별 (Male, Female)
- `SeniorCitizen`: 시니어 고객 여부 (1: Yes, 0: No)
- `Partner`: 배우자 여부 (Yes, No)
- `Dependents`: 부양 가족 여부 (Yes, No)
- `tenure`: 서비스 가입 기간 (개월 수)
- `PhoneService`: 전화 서비스 가입 여부 (Yes, No)
- `MultipleLines`: 다중 회선 가입 여부 (Yes, No, No phone service)
- `InternetService`: 인터넷 서비스 제공 유형 (DSL, Fiber optic, No)
- `OnlineSecurity`: 온라인 보안 서비스 가입 여부 (Yes, No, No internet service)
- `OnlineBackup`: 온라인 백업 서비스 가입 여부 (Yes, No, No internet service)
- `DeviceProtection`: 기기 보호 서비스 가입 여부 (Yes, No, No internet service)
- `TechSupport`: 기술 지원 서비스 가입 여부 (Yes, No, No internet service)
- `StreamingTV`: 스트리밍 TV 서비스 가입 여부 (Yes, No, No internet service)
- `StreamingMovies`: 스트리밍 영화 서비스 가입 여부 (Yes, No, No internet service)
- `Contract`: 계약 유형 (Month-to-month, One year, Two year)
- `PaperlessBilling`: 무지서 청구서 사용 여부 (Yes, No)
- `PaymentMethod`: 지불 방법 (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))
- `MonthlyCharges`: 월 요금
- `TotalCharges`: 총 요금
- `Churn`: 고객 이탈 여부 (Yes, No)

## 데이터셋 특징
- **범주형 변수**와 **연속형 변수**가 모두 포함되어 있습니다.
- 일부 변수에는 결측치가 존재하여 데이터 클렌징이 필요합니다.
- 고객 이탈 예측을 위한 다양한 특성이 포함되어 있어, 탐색적 데이터 분석(EDA) 및 머신러닝 모델링에 적합합니다.

## 탐색적 데이터 분석(EDA)
1. **기본 통계 확인**: 데이터의 분포와 기본 통계를 확인합니다.
2. **결측치 시각화**: 결측치의 패턴을 시각적으로 확인합니다.
3. **범주형 변수 분석**: 범주형 변수와 이탈 여부 간의 관계를 분석합니다.
4. **연속형 변수 분석**: 연속형 변수와 이탈 여부 간의 관계를 분석합니다.
5. **상관관계 분석**: 변수들 간의 상관관계를 분석하여 중요한 변수를 파악합니다.

Telco Customer Churn Dataset은 고객 이탈 예측을 위한 다양한 정보를 제공하는 데이터셋으로, 데이터 클렌징, 탐색적 데이터 분석 및 머신러닝 모델링에 매우 유용합니다. 이 데이터셋을 활용하여 고객 이탈의 주요 요인을 분석하고, 이를 바탕으로 효과적인 고객 유지 전략을 수립할 수 있습니다.

## 데이터클린징

### 문제 발생
결측치 확인을 위한 함수인 df.count(), df.isnull().sum()를 통해 확인했을 때 결측치가 존재하지 않는 것으로 보였습니다. 따라서 임의로 결측치를 생성하려고 하였으나 저의 보고서를 천천히 다시 검토하며 확인하던 중 문제가 있음을 발견했습니다. '데이터 보기'에서 총 요금 기준 정렬을 시행했을 때 몇몇개의 총 요금에 결측치가 존재했던 것입니다.

### 문제 원인
df.isnull().sum()을 통해 결측치를 확인할 때 아무것도 뜨지 않는다면, 이는 명시적인 결측치 값인 'NaN'이 아닌 다른 형태로 결측치가 저장되어 있을 가능성이 큽니다. 특히 TotalCharges와 같은 열에서 결측치가 문자열 형태로 저장되어 있을 수 있습니다. 예를 들어, 공백 문자열이나 특정 문자가 결측치를 의미할 수 있습니다. 실제로 무심코 지나쳤던 '데이터 보기'로 다시 돌아가서 확인한 결과 MonthlyCharges는 float64였으나 TotalCharges는 object였습니다.

### 문제 해결
문자열로 저장된 결측치를 명시적인 NaN 값으로 변환한 후 결측치를 다시 확인한 결과, 제대로 결측치가 측정이 되었습니다. 또 TotalCharges는 MonthlyCharges과 tenure을 곱한 값이어야 하기 때문에 그값이 명확하게 정해져 있습니다. 따라서 해당 결측치는 명백하게 실수로 발생한 것이라는 판단을 내릴 수 있었고 대체나 삭제로 해결하지 않고 total을 직접 계산해서 결측치에 해당하는 곳에 값을 삽입했습니다.


```python
df.count() # 데이터 갯수 확인
```

```python
df.isnull().sum() # 결측치 확인
```

```python
plt.figure(figsize=(12, 8))       # 결측치 시각화
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title('Original Data - Missing Data Heatmap')
plt.show()
```

```python
print(df['TotalCharges'].dtype) #데이터 타입 확인
```

```python
# 공백 문자열 확인
missing_total_charges = df[df['TotalCharges'].str.strip() == '']
print(missing_total_charges)

# 문자열 결측치의 개수 확인
print("결측치 개수:", len(missing_total_charges))
```

```python
# 공백 문자열을 NaN으로 변환
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan).astype(float)

# 결측치 확인
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 8))       # 결측치 시각화
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title('Original Data - Missing Data Heatmap')
plt.show()
```

```python
# TotalCharges의 결측치 계산 및 채우기
df['TotalCharges'] = df.apply(
    lambda row: row['MonthlyCharges'] * row['tenure'] if pd.isnull(row['TotalCharges']) else row['TotalCharges'], 
    axis=1
)

# 결측치가 채워진 후 확인
print("결측치 채운 후 확인:")
print(df.isnull().sum())
```

```python
# 결측치가 채워진 데이터 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title('Data after Filling Missing TotalCharges - Missing Data Heatmap')
plt.show()
```

## 데이터탐색

```python
# 데이터 바이닝 - 연속형 tenure 변수 범주화 cut() 함수 사용 
bins = [0, 24, 48, 72] # 1-20 미성년, 21-60 : 성년, 61-100 : 성년
df_tenure = pd.cut(df["tenure"], bins, labels =["Newbie", "Experienced", "VIP"])
df["tenure_class"] = df_tenure
df
```

```python
# tenure_class 변수의 분포 확인
tenure_class_distribution = df['tenure_class'].value_counts().sort_index()

# tenure_class 변수에 따른 이탈 여부 확인
churn_by_tenure_class = df.groupby('tenure_class', observed=True)['Churn'].value_counts(normalize=True).unstack().sort_index()

# 결과 출력
print("Tenure Class Distribution:")
print(tenure_class_distribution)

print("\nChurn Rate by Tenure Class:")
print(churn_by_tenure_class)
```

```python
# 계약 기간과 이탈 여부
# Contract 변수와 Churn 변수 간의 관계 탐색
contract_churn = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
print("\nContract 변수와 Churn 변수 간의 관계:")
print(contract_churn)
```

```python
# 개월 수와 이탈 여부
# tenure_class 변수와 Churn 변수 간의 관계 탐색
tenure_class_churn = df.groupby('tenure_class', observed=True)['Churn'].value_counts(normalize=True).unstack()
print("\ntenure_class 변수와 Churn 변수 간의 관계:")
print(tenure_class_churn)
```

```python
# 연령대와 월 요금, 이탈 여부 간의 관계 탐색
senior_monthly_churn = df.groupby(['SeniorCitizen', 'Churn'])['MonthlyCharges'].describe().unstack()
print("\nSeniorCitizen, Churn, MonthlyCharges 간의 관계:")
print(senior_monthly_churn)
```

```python
# 계약 유형과 인터넷 서비스 유형, 이탈 여부 간의 관계 탐색
contract_churn = df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
print("\nContract 변수와 Churn 변수 간의 관계:")
print(contract_churn)

internet_service_churn = df.groupby('InternetService')['Churn'].value_counts(normalize=True).unstack()
print("\nInternetService 변수와 Churn 변수 간의 관계:")
print(internet_service_churn)
```

```python
# 결제 방법과 월 요금, 이탈 여부 간의 관계 탐색
payment_monthly_churn = df.groupby(['PaymentMethod', 'Churn'])['MonthlyCharges'].describe().unstack()
print("\nPaymentMethod, Churn, MonthlyCharges 간의 관계:")
print(payment_monthly_churn)
```

## 데이터 또는 분석 과정 차별화 포인트

### 1. 데이터 전처리의 철저함
데이터 전처리는 분석의 기초이자 가장 중요한 단계 중 하나입니다. 원본 데이터를 정확히 이해하고, 결측치를 처리하며, 데이터 타입을 변환하는 과정은 분석의 신뢰성을 높이는 데 필수적입니다.
- **결측치 처리**: 이번 분석에서는 `TotalCharges` 열에 존재하는 결측치를 `MonthlyCharges`와 `tenure`를 곱하여 계산하는 방식으로 대체하였습니다. 이는 결측치가 있는 데이터를 단순히 제거하는 것이 아니라, 다른 관련 데이터를 활용하여 합리적으로 보완한 점에서 차별화됩니다.
- **데이터 타입 변환**: `TotalCharges` 열의 데이터를 수치형으로 변환하고, `Churn` 열을 범주형으로 변환하여 데이터 분석과 시각화 과정에서 발생할 수 있는 오류를 사전에 방지하였습니다.

### 2. 탐색적 데이터 분석 (EDA)의 철저함
탐색적 데이터 분석(EDA)은 데이터를 깊이 이해하고, 숨겨진 패턴을 발견하며, 분석의 방향성을 설정하는 데 매우 중요합니다.
- **범주형 변수의 값 분포 확인**: 각 범주형 변수에 대해 값의 분포를 확인함으로써 데이터의 전반적인 특성을 파악했습니다. 이를 통해 각 변수의 대표적인 값과 이상치를 쉽게 식별할 수 있었습니다.
- **연속형 변수의 기초 통계량 분석**: `tenure`, `MonthlyCharges`, `TotalCharges`와 같은 연속형 변수의 기초 통계량을 분석하여 데이터의 분포, 중앙값, 평균 등을 확인했습니다. 이를 통해 데이터의 전반적인 경향성을 파악할 수 있었습니다.

### 3. 다변량 탐색
단일 변수나 이변량 탐색을 넘어서, 다변량 탐색을 통해 보다 복잡한 관계를 이해하고 데이터의 전반적인 구조를 파악하는 데 중점을 두었습니다.
- **연령대와 월 요금, 이탈 여부 간의 관계**: `SeniorCitizen`, `MonthlyCharges`, `Churn` 변수를 사용하여 연령대가 월 요금과 이탈 여부에 미치는 영향을 탐색했습니다. 이를 통해 연령에 따른 요금 패턴과 이탈 패턴을 이해할 수 있었습니다.
- **계약 유형과 인터넷 서비스 유형, 이탈 여부 간의 관계**: `Contract`, `InternetService`, `Churn` 변수를 사용하여 계약 유형과 인터넷 서비스 유형이 이탈 여부에 미치는 영향을 분석했습니다. 이를 통해 다양한 서비스 조합이 이탈률에 미치는 영향을 파악할 수 있었습니다.
- **결제 방법과 월 요금, 이탈 여부 간의 관계**: `PaymentMethod`, `MonthlyCharges`, `Churn` 변수를 사용하여 결제 방법이 월 요금과 이탈 여부에 미치는 영향을 탐색했습니다. 이를 통해 결제 방법에 따른 고객의 요금 지불 패턴과 이탈 패턴을 이해할 수 있었습니다.

### 4. 탐색적 데이터 분석(EDA)의 시각화
EDA의 시각화는 데이터를 이해하고 해석하는 데 중요한 역할을 합니다. 시각화를 통해 데이터의 분포와 변수 간의 관계를 직관적으로 파악할 수 있습니다.
- **`tenure_class` 변수의 분포 시각화**: `tenure_class` 변수의 값을 Countplot을 사용하여 시각화함으로써 각 클래스의 고객 수를 명확히 파악할 수 있었습니다.
- **`tenure_class` 변수에 따른 이탈 여부 시각화**: `tenure_class` 변수에 따른 이탈 여부를 Stacked Bar Plot으로 시각화하여 각 클래스에서의 이탈률을 명확히 비교할 수 있었습니다.
- **계약 기간과 이탈 여부 시각화**: 계약 유형별 이탈률을 Stacked Bar Plot으로 시각화하여 어떤 계약 유형에서 이탈률이 높은지를 한눈에 파악할 수 있었습니다.
- **개월 수와 이탈 여부 시각화**: `tenure_class` 변수와 이탈 여부를 Stacked Bar Plot으로 시각화하여 각 클래스의 이탈 패턴을 명확히 이해할 수 있었습니다.
